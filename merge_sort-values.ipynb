```python
import pandas as pd
```

```python
nufus = pd.read_csv("data/tr_iller.csv")
```

```python
nufus.head()
```

```python
kurban = pd.read_csv("data/kurban_kaza.csv", names=["iller","kazalar"])
```

```python
kurban.head()
```

```python
bilanco = pd.merge(nufus, kurban, left_on="İl",right_on="iller")
bilanco.drop(["iller","Yıl","Erkek Nüfusu","Kadın Nüfusu","Nüfus Yüzdesi"],axis=1, inplace=True)
```

```python
bilanco.sort_values("kazalar", ascending=False).head()
```

```python
bilanco.dtypes
```

```python
bilanco["Toplam Nüfus"] = bilanco["Toplam Nüfus"].str.replace(".","")
```

```python
bilanco.dtypes
```

```python
bilanco["Toplam Nüfus"] = bilanco["Toplam Nüfus"].astype("int")
```

```python
bilanco.dtypes
```

```python
bilanco["oran-10binde"] = bilanco.kazalar / bilanco["Toplam Nüfus"] * 10000
```

```python
bilanco.sort_values("oran-10binde",ascending=False)
```
